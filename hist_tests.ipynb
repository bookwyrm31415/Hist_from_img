```python
import matplotlib.pyplot as plt
from skimage import io

%matplotlib inline
```

```python
image = io.imread('b_w_dog.jpeg')
```

```python
hist = plt.hist(image.ravel(), bins = 256)
plt.show()
```

```python
image_2 = io.imread('color_dog.jpeg')
```

```python
hist = plt.hist(image_2.ravel(), bins = 256, color = 'orange', )
hist = plt.hist(image_2[:, :, 0].ravel(), bins = 256, color = 'red', alpha = 0.5)
hist = plt.hist(image_2[:, :, 1].ravel(), bins = 256, color = 'Green', alpha = 0.5)
hist = plt.hist(image_2[:, :, 2].ravel(), bins = 256, color = 'Blue', alpha = 0.5)
hist = plt.xlabel('Intensity Value')
hist = plt.ylabel('Count')
hist = plt.legend(['Total', 'Red_Channel', 'Green_Channel', 'Blue_Channel'])
plt.show()
```
